```python
import cv2
import pytesseract
```

```python
img1=cv2.imread('ocr1.jpeg')
img1
```

```python
cv2.imshow('OCR',img1)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
#set path tesseract ocr
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
#text extract
data=pytesseract.image_to_string(img1)
data
```

```python
#properties in form of dictionary
dic=pytesseract.image_to_data(img1,output_type=pytesseract.Output.DICT)
dic
```

```python
#to draw rectangle box
#left:43,top:75,width:210,height:32 not need for for loop here as only 1 image is there
x,y,w,h=43,75,210,32
cv2.rectangle(img1,(x,y),(x+w,y+h),(0,0,255),4)
cv2.imshow('Image',img1)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
